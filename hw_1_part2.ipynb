```
def simulated_annealing(initial, generate_f, energy_f, T = 2 , alpha = 0.95, iters = 50, min_T = 0.25):
  best = initial
  e_best = energy_f(initial)
  x0 = initial
  history = [initial]
  while T > min_T:
    for i in range(iters):
      x = generate_f(x0)
      e_x = energy_f(x)
      e_x0 = energy_f(x0)
      if(e_x < e_x0):
        x0 = x
        if e_x < e_best:
          e_best = e_x
          best = x
      else:
        accept_ratio = np.exp(-e_x/T)/ np.exp(-e_x0/T)
        if accept_ratio < np.random.uniform(0.0,1.0):
          x0 = x
    T = T * alpha
    history.append(best)
  return best,e_best,history
```

```
from geopy.distance import distance
dis = None
def store_distance(n = 30): # preprocess distance between cities
  global dis
  dis = [[0]*n for i in range(n)]
  for i in range(n):
    for j in range(n):
      dis[i][j] = distance((values[i][2],values[i][3]),(values[j][2],values[j][3])).km
  print(dis)
store_distance()
```

```
def energy_fun(order): # sum of distance
  n = len(order)
  cost = 0
  for i in range(1,n):
    cost += dis[order[i-1]][order[i]]
  return cost
```

```
def generate_next(order): # mutation to order
  n = len(order)
  order = order.copy()
  i = np.random.randint(n)
  j = np.random.randint(n)
  order[i],order[j] = order[j],order[i]
  return order
import random
def build_initial(n = 30): # generate random permutation
  order = []
  for i in range(n):
    order.append(i)
  random.shuffle(order)
  return order
```

```
# run 1
initial = build_initial()
T = 2000
alpha = 0.99
T_min = 2
order,cost,history = simulated_annealing(initial=initial,generate_f=generate_next, energy_f=energy_fun,T=T,alpha=alpha,min_T = T_min)
cost
```
